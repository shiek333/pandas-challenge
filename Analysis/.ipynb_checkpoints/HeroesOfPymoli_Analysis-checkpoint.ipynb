# Heroes of Pymoli Analysis

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "..\Resources\purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

```python
total_players = pd.DataFrame({"Total Players":[purchase_data['SN'].nunique()]})
total_players
```

## Purchasing Analysis (Total)

```python
totalRevenue = purchase_data['Price'].sum()
purchasing_analysis_total = pd.DataFrame({"Number of Unique Items":[purchase_data['Item Name'].nunique()],
                                        "Average Price":[f"${round(purchase_data['Price'].mean(),2)}"],
                                       "Number of Purchases":[purchase_data['Purchase ID'].count()],
                                       "Total Revenue":["${:,.2f}".format(totalRevenue)]}) 
purchasing_analysis_total
```

## Gender Demographics

```python
# Gender Demographics Variables
male = purchase_data.loc[purchase_data['Gender']=='Male']
maleCount = male['SN'].nunique()
female = purchase_data.loc[purchase_data['Gender']=='Female']
femaleCount = female['SN'].nunique()
otherCount = purchase_data['Gender'].loc[purchase_data['Gender']=='Other / Non-Disclosed'].count()
genderTotal = purchase_data['Gender'].count()
malePercent = round(maleCount/genderTotal,4)*100
femalePercent = round(femaleCount/genderTotal,4)*100
otherPercent = round(otherCount/genderTotal,4)*100
gender_list = ['Male','Female','Other / Non-Disclosed']

# Gender Demographics DataFrame
gender_demographics = pd.DataFrame({"Total Count":[maleCount,femaleCount,otherCount],
                                   "Percentage of Players":[malePercent,
                                                            femalePercent,
                                                            otherPercent]})
gender_demographics['Percentage of Players'] = gender_demographics['Percentage of Players'].map("{:,.2f}%".format)
gender_demographics = gender_demographics.set_index([gender_list])
gender_demographics
```


## Purchasing Analysis (Gender)

```python
pag_group = purchase_data.groupby(['Gender'])
genderTotal = purchase_data.groupby(['Gender'])['SN'].nunique()
purchasing_analysis_gender = pd.DataFrame(purchase_data.groupby(['Gender'])['Purchase ID'].count())
purchasing_analysis_gender = purchasing_analysis_gender.rename(columns={'Purchase ID':"Purchase Count"})
purchasing_analysis_gender['Average Purchase Price'] = pag_group['Price'].mean().map("${:,.2f}".format)
purchasing_analysis_gender['Total Purchase Value'] = pag_group['Price'].sum().map("${:,.2f}".format)
purchasing_analysis_gender['Avg Total Purchase per Person'] = ((pag_group['Price'].mean()*purchasing_analysis_gender['Purchase Count'])/genderTotal).map("${:.2f}".format)
purchasing_analysis_gender
```

## Age Demographics

```python
bins = [0,9,14,19,24,29,34,39,100]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data['Age Category'] = pd.DataFrame(pd.cut(purchase_data['Age'], bins=bins, labels=labels))
age_demo = purchase_data.groupby(['Age Category'])['SN'].nunique().to_frame()
age_demo = age_demo.rename(columns={"SN":"Total Count"})
age_sum = purchase_data['SN'].nunique()
age_demo['Percentage of Players'] = ((age_demo['Total Count']/age_sum)*100).map("{:.2f}%".format)
age_demo = age_demo.reindex(labels)
age_demo.index.name = None
age_demo
```

## Purchasing Analysis (Age)

```python
bins = [0,9,14,19,24,29,34,39,100]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
age_demo_group = purchase_data.groupby(['Age Category'])
ageTotal = purchase_data.groupby(['Age Category'])['SN'].nunique()
pa_age = pd.DataFrame([])
pa_age = pa_age.reindex(labels)
pa_age['Purchase Count'] = age_demo_group['Purchase ID'].count()
pa_age['Average Purchase Price'] = age_demo_group['Price'].mean().map("${:,.2f}".format)
pa_age['Total Purchase Value'] = age_demo_group['Price'].sum().map("${:,.2f}".format)
pa_age['Avg Total Purchase per Person'] = ((age_demo_group['Price'].mean()*pa_age['Purchase Count'])/ageTotal).map("${:.2f}".format)
pa_age
```

## Top Spenders

```python
topSpenders_group = purchase_data.groupby(['SN'])
topSpenders = pd.DataFrame([])
topSpenders['Purchase Count'] = topSpenders_group['Purchase ID'].count()
topSpenders['Average Purchase Price'] = topSpenders_group['Price'].mean()
topSpenders['Total Purchase Value'] = topSpenders['Purchase Count'] * topSpenders['Average Purchase Price']
topSpenders = topSpenders.sort_values(by='Total Purchase Value',ascending=False)
topSpenders['Average Purchase Price'] = topSpenders['Average Purchase Price'].map("${:,.2f}".format)
topSpenders['Total Purchase Value'] = topSpenders['Total Purchase Value'].map("${:,.2f}".format)
topSpenders.head(5)
```

## Most Popular Items

```python
popItems_group = purchase_data.groupby(['Item ID','Item Name'])
popItems = pd.DataFrame([])
popItems['Purchase Count'] = popItems_group['Purchase ID'].count()
popItems['Item Price'] = popItems_group['Price'].mean()
popItems['Total Purchase Value'] = popItems['Purchase Count'] * popItems['Item Price']
popItems = popItems.sort_values(by=['Purchase Count'],ascending=False)
popItems['Item Price'] = popItems['Item Price'].map("${:,.2f}".format)
popItems['Total Purchase Value'] = popItems['Total Purchase Value'].map("${:,.2f}".format)
popItems.head(5)
```

## Most Profitable Items

```python
profitItems_group = purchase_data.groupby(['Item ID','Item Name'])
profitItems = pd.DataFrame([])
profitItems['Purchase Count'] = profitItems_group['Purchase ID'].count()
profitItems['Item Price'] = profitItems_group['Price'].mean()
profitItems['Total Purchase Value'] = profitItems['Purchase Count'] * profitItems['Item Price']
profitItems = profitItems.sort_values(by=['Total Purchase Value'],ascending=False)
profitItems['Item Price'] = profitItems['Item Price'].map("${:,.2f}".format)
profitItems['Total Purchase Value'] = profitItems['Total Purchase Value'].map("${:,.2f}".format)
profitItems.head(5)
```
